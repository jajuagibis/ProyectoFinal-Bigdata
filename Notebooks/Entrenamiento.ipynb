## Código implementado para plotear las imágenes y tomar los puntos para el entrenamiento

```python
#Importar modulos
import os
import numpy as np
from skimage.io import imread
import plotly.express as px
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Montar Google Drive
from google.colab import drive
drive.mount('/content/gdrive')
```

```python

# Definir la ruta de la carpeta de imágenes en tu Google Drive
carpeta_imagenes = '/content/gdrive/MyDrive/Bigdata&Bioinformática - Notebook/Proyecto Final /Sanas'

# Obtener lista de imágenes válidas
archivos_imagenes = [f for f in os.listdir(carpeta_imagenes) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Seleccionar 30 imágenes al azar

imagenes_seleccionadas = np.random.choice(archivos_imagenes, size=min(5, len(archivos_imagenes)), replace=False)

# Mostrar cada imagen con Plotly
for nombre_imagen in imagenes_seleccionadas:
    ruta_imagen = os.path.join(carpeta_imagenes, nombre_imagen)
    imagen = imread(ruta_imagen)
    imagen_rgb = imagen[..., :3]  # Asegurar solo RGB

    # Mostrar imagen interactiva
    fig = px.imshow(imagen_rgb, title=nombre_imagen)
    fig.update_layout(coloraxis_showscale=False)
    fig.show()
```
